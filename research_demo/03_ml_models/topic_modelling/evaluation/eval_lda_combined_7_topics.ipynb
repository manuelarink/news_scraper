# Evaluation of LDA model for prediction of news category with 7 topics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load results of topic modelling with LDA model combining title, description and text

```python
pred = pd.read_csv('train_lda_combined_7.csv')
```

```python
pred.head()
```

```python
pred.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
pred.head()
```

## Visualizations

### Distribution of topics and categories

```python
sns.histplot(pred['topic'])
plt.title('Histogram of news topics found by Latent Dirichlet Allocation on title, description and text and 7 topics')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=pred, x='topic', hue='source')
plt.title('Histogram of news topics found by Latent Dirichlet Allocation on title, description and text and 7 topics')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=pred, x='topic', hue='source', fill=True)
plt.title('Histogram of news topics found by Latent Dirichlet Allocation on title, description and text and 7 topics')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=pred, x='category', hue='source')
plt.title('Histogram of prelabeled categories')
plt.show()
```
